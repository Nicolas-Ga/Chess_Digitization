# Chess Dataset

```python
images_zip = tf.keras.utils.get_file(
    'chess_training1.zip',  "https://github.com/Nicolas-Ga/Chess_Digitization/blob/main/chess_training1.zip?raw=true",
    cache_dir='.', cache_subdir='chess_training',
    extract=True)
images_zip = tf.keras.utils.get_file(
    'chess_training_2.zip',  "https://github.com/Nicolas-Ga/Chess_Digitization/blob/main/chess_training_2.zip?raw=true",
    cache_dir='.', cache_subdir='chess_training',
    extract=True)
images_zip = tf.keras.utils.get_file(
    'chess_testing.zip',  "https://github.com/Nicolas-Ga/Chess_Digitization/blob/main/chess_testing.zip?raw=true",
    cache_dir='.', cache_subdir='chess_testing',
    extract=True)
newpath = '/content/gray_scale'
if not os.path.exists(newpath):
    os.makedirs(newpath)
newpath = '/content/dataset'
if not os.path.exists(newpath):
    os.makedirs(newpath)
newpath = '/content/testing'
if not os.path.exists(newpath):
    os.makedirs(newpath)
```

```python
dataset_size = 71;
img_size = 256, 256
# create a ZipFile object
zipObj = ZipFile('/content/dataset/dataset.zip', 'w')


for i in range(0, dataset_size):
  name = "/content/chess_training/" + str(i) + ".jpg"
  image = cv.imread(name)
  gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
  name = "/content/gray_scale/" + str(i) + ".jpg" 
  cv.imwrite(name,gray)
  im = Image.open(name)
  im_resized = im.resize(img_size, Image.ANTIALIAS)
  name = "/content/dataset/" + str(i) + ".jpg"
  im_resized.save(name)
  # Add multiple files to the zip
  zipObj.write('/content/dataset/' + str(i) + '.jpg')
# close the Zip File
zipObj.close()
```

```python
testing_size = 31;
# create a ZipFile object
zipObj = ZipFile('/content/testing/testing.zip', 'w')


for i in range(dataset_size, dataset_size+testing_size):
  name = "/content/chess_testing/" + str(i) + ".jpg"
  image = cv.imread(name)
  gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
  name = "/content/gray_scale/" + str(i) + ".jpg" 
  cv.imwrite(name,gray)
  im = Image.open(name)
  im_resized = im.resize(img_size, Image.ANTIALIAS)
  name = "/content/testing/" + str(i) + ".jpg"
  im_resized.save(name)
  # Add multiple files to the zip
  zipObj.write('/content/testing/' + str(i) + '.jpg')
# close the Zip File
zipObj.close()
```
